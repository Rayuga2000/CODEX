```python
import sqlite3 as sq3
import pandas as pd
```

```python
path = 'Asset_Study/database/classic_rock.db'
con = sq3.Connection(path)
```

```python
con
```

```python
query='''SELECT * FROM rock_songs'''
observation=pd.read_sql(query,con)
display(observation)
```

```python
query='''
SELECT Artist, Release_Year, COUNT(*) AS num_songs, AVG(PlayCount) AS avg_plays  
    FROM rock_songs
    GROUP BY Artist, Release_Year
    ORDER BY num_songs desc;
'''
i=1
# Execute the query
observation_generator = pd.read_sql(query,
                            con,
                            parse_dates=['Release_Year'], # Parse `Release_Year` as a date
                            chunksize=5 # Allows for streaming results as a series of shorter tables
                           )
#print(type(observation_generator))
for x in observation_generator:
    print("Observation ",i,":\n",x,"\n")
    i+=1
```
