```python
import numpy as np

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, learning_rate):
        # Initialize weights and biases
        self.weights_input_hidden = np.random.rand(input_size, hidden_size)
        self.bias_hidden = np.zeros((1, hidden_size))
        self.weights_hidden_output = np.random.rand(hidden_size, output_size)
        self.bias_output = np.zeros((1, output_size))

        # Set learning rate
        self.learning_rate = learning_rate

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def forward(self, inputs):
        # Forward pass
        self.hidden_layer_input = np.dot(inputs, self.weights_input_hidden) + self.bias_hidden
        print(self.hidden_layer_input)
        self.hidden_layer_output = self.sigmoid(self.hidden_layer_input)

        self.output_layer_input = np.dot(self.hidden_layer_output, self.weights_hidden_output) + self.bias_output
        self.predicted_output = self.sigmoid(self.output_layer_input)

        return self.predicted_output

    def backward(self, inputs, target):
        # Backward pass
        error = target - self.predicted_output

        # Output layer gradients
        output_delta = error * self.sigmoid_derivative(self.predicted_output)
        self.weights_hidden_output += self.learning_rate * np.dot(self.hidden_layer_output.T, output_delta)
        self.bias_output += self.learning_rate * np.sum(output_delta, axis=0, keepdims=True)

        # Hidden layer gradients
        hidden_error = np.dot(output_delta, self.weights_hidden_output.T)
        hidden_delta = hidden_error * self.sigmoid_derivative(self.hidden_layer_output)
        self.weights_input_hidden += self.learning_rate * np.dot(inputs.T, hidden_delta)
        self.bias_hidden += self.learning_rate * np.sum(hidden_delta, axis=0, keepdims=True)

    def train(self, inputs, targets, epochs):
        for epoch in range(epochs):
            for i in range(len(inputs)):
                # Forward pass
                predicted_output = self.forward(inputs[i:i+1])

                # Backward pass
                self.backward(inputs[i:i+1], targets[i:i+1])

# Example usage:
input_size = 2
hidden_size = 4
output_size = 1
learning_rate = 0.1

# Sample training data
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
targets = np.array([[0], [1], [1], [0]])

# Create and train the neural network
neural_network = NeuralNetwork(input_size, hidden_size, output_size, learning_rate)
neural_network.train(inputs, targets, epochs=10000)

# Test the trained network
test_input = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
predicted_output = neural_network.forward(test_input)

print("\nFinal Predictions:")
for i in range(len(test_input)):
    print(f"Input: {test_input[i]}, Predicted Output: {predicted_output[i]}")
```
